```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

path="C:\\Users\\SENTHIL\\Desktop\\Training\\"
df_file =pd.read_csv(path+"BigMartSalesData.csv")
print(df_file)
```

```python
df_2011 = df_file[df_file['Year'] == 2011]
df_2011 = df_file[['Month','Amount']].groupby('Month',as_index=False) 
df_2011_sales=df_2011.first()

df_2011_low_sales = df_2011_sales[df_2011_sales['Amount']==df_2011_sales.Amount.min()]
df_2011_low_sales = df_2011_low_sales['Month'].to_string(index=False)
print("Lowest sales month",df_2011_low_sales)
```

```python
print(df_2011_sales)
```

```python
plt.plot(df_2011_sales['Month'], df_2011_sales['Amount'])
plt.xlabel('Month')
plt.ylabel('Sales Amt')
plt.grid(True)
plt.title("Total Sales Per Month for Year 2011")
plt.legend(['Month'])
plt.show()
```

```python
plt.bar(df_2011_sales['Month'], df_2011_sales['Amount'])
plt.xlabel('Month')
plt.ylabel('Sales Amt')
plt.grid(True)
plt.title("Total Sales Per Month for Year 2011")
plt.legend(['Month'])
plt.show()
```

```python
df_2011_country_wise = df_file[df_file['Year'] == 2011]
df_2011_country_wise = df_2011_country_wise[["Country","Amount"]].groupby('Country',as_index=False)
df_2011_country_wise=df_2011_country_wise.first()

df_2011_country_wise["Country"]=np.where((df_2011_country_wise.Amount<=45),"Others",df_2011_country_wise.Country)
df_2011_country_wise = df_2011_country_wise[["Country","Amount"]].groupby('Country',as_index=False)
df_2011_country_wise=df_2011_country_wise.first()


print(df_2011_country_wise.sort_values(by=['Amount'], ascending=False))

sales=np.array(df_2011_country_wise["Amount"])
labels=df_2011_country_wise["Country"]
plt.pie(sales,labels=labels,shadow=True, startangle=50)
plt.title("Chart for Year 2011 Country Wise")
#plt.legend(labels)
plt.figure(figsize=(10,10))
plt.show()
```
