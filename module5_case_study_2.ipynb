```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salaries.csv')
df
```

```python
"""
Compute how much total salary cost has increased from year 2011 to 2014
"""
pay=df[['Year','TotalPay']]
pay.groupby('Year').mean()
```

```python
"""
Which Job Title in Year 2014 has highest mean salary?
"""
df_2014 = df[df['Year']==2014]
df_2014=df_2014[['JobTitle','TotalPay']]
df_2014_mean =df_2014.groupby('JobTitle',as_index=False).mean() 
df_2014_mean.max()
```

```python
"""
3. How much money could have been saved in Year 2014 by stopping
OverTimePay?
"""
df_2014 = df[df['Year']==2014]
pay=df_2014[['Year','OvertimePay']]
pay=pay.groupby('Year',as_index=False).sum()
round(pay)
```

```python
"""
Which are the top 5 common job in Year 2014 and how much do they cost SFO
?

"""
df_2014 = df[df['Year']==2014]
pay=df_2014[['JobTitle','TotalPayBenefits']]
pay.nlargest(5,'TotalPayBenefits')
```

```python
"""

Who was the top earning employee across all the years?

"""

df_2014 = df[df['Year']==2014]
pay=df_2014[['JobTitle','TotalPayBenefits']]
pay
```
