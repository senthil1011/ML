```python
"""You are given a dataset, which is present in the LMS, containing the number of hurricanes 
occurring in the United States along the coast of the Atlantic. Load the data from the dataset 
into your program and plot a Bar Graph of the data, taking the Year as the x-axis and the number 
of hurricanes occurring as the Y-axis."""

import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
path = "C:\\Users\\SENTS-IN\\Desktop\\Training\\"
df_hurricanes = pd.read_csv(path+"hurricanes.csv")
print(df_hurricanes)
```

```python
x=df_hurricanes["Year"]
y=df_hurricanes["Hurricanes"]
plt.bar(x,y)
plt.xlabel("Year")
plt.ylabel("Hurricanes")
plt.title("No.of hurricanes occurred in US")
plt.show()
```

```python
"""
The dataset given, records data of city temperatures over the years’2014 and 2015. Plot the histogram of the 
temperatures over this period for the cities of San Francisco and Moscow."""
df_city_temp=pd.read_csv(path+"CityTemps.csv")
print(df_city_temp)
```

```python
df_temp_Moscow = df_city_temp["Moscow"]
df_temp_Melbourne = df_city_temp["Melbourne"]
df_temp_San = df_city_temp["San Francisco"]
plt.hist([df_temp_Moscow,df_temp_San],bins=[-10,-5,0,5,10,15,20],rwidth=0.95,color=['g','orange'],label =['Moscow','San Fransisco'])
plt.legend()
plt.show()

```

```python
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]
y_error = [0.12, 0.13, 0.2, 0.1]
plt.figure(figsize=(4,5))
plt.plot(x,y,color="green",marker="+",linestyle="--")
plt.xlabel("x",fontsize = '14')
plt.ylabel("y",fontsize = '14')
plt.title("X vs Y")
plt.show()

plt.figure(figsize=(4,5))
plt.plot(x,y_error,color="blue",marker="+",linestyle="--")
plt.xlabel("x",fontsize = '14')
plt.ylabel("y Error",fontsize = '14')
plt.title("X vs Y Error")
plt.show()
plt.savefig('output.jpeg', dpi=1200)
```
